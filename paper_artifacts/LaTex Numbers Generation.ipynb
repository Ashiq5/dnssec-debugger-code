```python
import json
import pandas as pd

def beautiful_number(value):
    str_value = ""
    len_value = len(value)

    for i, character in enumerate(value):
        str_value += character
        if i != len_value-1 and (i+1) % 3 == len_value % 3 :
            str_value += ","

    return str_value



class dynamicNumbers():
    def __init__(self, command_name, value, is_number = False, is_percentage = False):
        self.command_name = command_name
        self.value = str(value)
        self.is_percentage = is_percentage
        if is_number:


            self.value = beautiful_number(self.value)

    def __str__(self):
        # without coloring
        # return f'\\newcommand{{\\DfixerDynamic{self.command_name}}}{{{self.value}{"\\%" if self.is_percentage else ""}}}\n'
        return f'\\newcommand{{\\DfixerDynamic{self.command_name}}}{{{self.value}{"\\%" if self.is_percentage else ""}}}\n'

```

```python
with open("generated/LaTexData.json", 'r') as fp:
    latex_export = json.load(fp)

nb_line = latex_export.get("nb_line")
nb_files_with_errors = latex_export.get("nb_files_with_errors")
nb_were_we_generated_errors = latex_export.get("nb_were_we_generated_errors")
nb_where_generated_errors_covers_intended = latex_export.get("nb_where_generated_errors_covers_intended")
nb_files_with_errors_nnsic = latex_export.get("nb_files_with_errors_nnsic")
nb_were_we_generated_errors_nnsic = latex_export.get("nb_were_we_generated_errors_nnsic")
nb_where_generated_errors_covers_intended_nnsic = latex_export.get("nb_where_generated_errors_covers_intended_nnsic")
errors_type_encountered = latex_export.get("errors_type_encountered")
errors_type_generated = latex_export.get("errors_type_generated")
case_cannot_be_fixed = latex_export.get("case_cannot_be_fixed")
case_cannot_be_fixed_without_nzic = latex_export.get("case_cannot_be_fixed_without_nzic")
errors_combination = latex_export.get("errors_combination")
errors_combination_fully_covered = latex_export.get("errors_combination_fully_covered")
errors_combination_covered = latex_export.get("errors_combination_covered")
nb_of_snapshot_sign_the_zone_first_iteration = latex_export.get("nb_of_snapshot_sign_the_zone_first_iteration")
total_combination_of_errors_in_the_signature_first_step = latex_export.get("total_combination_of_errors_in_the_signature_first_step")
pct_sign_first_iter = latex_export.get("pct_sign_first_iter")
pct_sign_second_iter = latex_export.get("pct_sign_second_iter")
pct_sign_third_iter = latex_export.get("pct_sign_third_iter")
pct_removeds_first_iter = latex_export.get("pct_removeds_first_iter")
pct_removeds_second_iter = latex_export.get("pct_removeds_second_iter")
pct_removeds_third_iter = latex_export.get("pct_removeds_third_iter")
```
